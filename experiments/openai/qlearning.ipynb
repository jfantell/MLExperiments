## Reinforcement Learning with OpenAI Gym

This notebook is based on the tutorial [Reinforcement Q-Learning from Scratch in Python with OpenAI Gym](https://www.learndatasci.com/tutorials/reinforcement-q-learning-scratch-python-openai-gym/). This notebook can be used to train and evaluate OpenAI RL problems with a discrete state/observation spaces and discrete action spaces.

```python
%matplotlib inline
import gym
import numpy as np
import toolkit.utils as utils
import matplotlib.pyplot as plt
```

```python
# Create environment
env = gym.make("Taxi-v3", render_mode = "human", new_step_api=True)
env.reset()
```

```python
# Explore state and action space
print("Action Space {}".format(env.action_space))
print("State Space {}".format(env.observation_space))
```

```python
# Setup Q-Learning table (state x action matrix)
q_table = np.zeros([env.observation_space.n, env.action_space.n])
```

```python
# Train the agent
utils.train(env, set([-10]), q_table)
```

```python
# Evaluate Agent: Brute Force Method
total_epochs, total_penalties = utils.evaluate(env, set([-10]))
```

```python
# Evaluate Agent: Q Learning)
utils.evaluate(env, set([-10]), q_table=q_table)
```
